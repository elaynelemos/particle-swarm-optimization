## Superfície alvo:

#### Função matemática
$f(x, y) = 5 - 2 \times \sin(3 \times (\frac{(\frac{x}{4})^2}\pi + \frac{(\frac{y}{5})^2}\pi)$

#### Plotando a função escolhida

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  
import numpy as np
import random
```

```python
def math_function(x, y):
    return 5 - 2*np.sin(3 * ((x/4)**2/np.pi + (y/5)**2/np.pi))
```

```python
def plot_max_point(x: list, y: list, plotable):
    x_max = y_max = z_max = -float('inf')
    for (v1, v2) in [(a,b) for a in x for b in y]:
        z = math_function(v1, v2)
        if z > z_max:
            x_max = v1
            y_max = v2
            z_max = z

    plotable.scatter3D(x_max, y_max, z_max, s=500, color='g')
    print(f'Max point: f({x_max}, {y_max}) = {z_max}')
```

```python
def plot_function():
    x = y = np.arange(-5, 5, 0.025)
    X, Y = np.meshgrid(x, y)
    zs = np.array(math_function(np.ravel(X), np.ravel(Y)))
    Z = zs.reshape(X.shape)

    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z, cmap='jet')

    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('f(x, y)')
    ax.set_title('Surface')

    plot_max_point(x, y, ax)
    plt.show()
```

```python
plot_function()
```
