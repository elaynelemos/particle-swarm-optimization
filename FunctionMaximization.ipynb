```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  
import numpy as np
import random
```

### SuperfÃ­cie alvo:

```python
def math_function(x, y):
    return -np.sin(5 * (x**2/121*np.pi + y**2/121*np.pi))
```

```python
def plot_function():
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111, projection='3d')
    x = y = np.arange(-3, 3, 0.025)
    X, Y = np.meshgrid(x, y)
    zs = np.array(math_function(np.ravel(X), np.ravel(Y)))
    Z = zs.reshape(X.shape)
    
    ax.plot_surface(X, Y, Z, cmap='jet')
    x_max = y_max = z_max = -float('inf')
    for v1 in x:
        for v2 in y:
            z = math_function(v1, v2)
            if z > z_max:
                x_max = v1
                y_max = v2
                z_max = z 
    ax.scatter3D(x_max, y_max, z_max, s=500, color='g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('f(x, y)')
    ax.set_title('Sin function')
    
    print(f'Max point: f({x_max}, {y_max}) = {z_max}')

    plt.show()
```

```python
plot_function()
```
